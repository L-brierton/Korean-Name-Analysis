# Introduction
I've always been fascinated by different languages, particularly those with vastly different ways of conceptionilising what we would consider fundamental ideas. One of these is names. I found myself interested in East Asian languages, particularly Korean, Japanese and varieties of Chinese. I enjoyed their vastly different writing systems and unique liguistic phonomenon, such as tones in Chinese and adjectives conjugating like verbs in Korean. As I translated sentences back and forth to get a feel for the laguage, I noticed some oddities pertaining to names. Names could not be properly gendered in these languages and often led to the incorrect English pronoun being used. This is because, unlike names in English and many other European languages, there is no strict masculine/feminine name endings. In most cases in these languages, the gender is dictated by the characters used to spell it. Despite some misinformation, names in these languages do tend to skew towards one gender or another based on sound or the underlying character meaning. I wished to develop a way that a machine could determine the gender of names.

I chose Korean for this task as while they do use characters for names, more often then not, these characters (called Hanja (한자) when used to write Korean) are only used in official documentation and such. In most other contexts, names are spelt using Hangul (한글), an alphabetic writing system. For example, the name Kim So-hyun (where Kim is the surname and SO-hyun is the given name), would be written as 김소현 in Hangul and 金所炫 could be a possible Hanja combination. This means far more emphasis is put on the sound of the syllables rather than the characters. I thought then this would make a good prototype for my tool.

```python
def convert_hangul(x):
    hangul_string = str(x.encode('utf-8'))
    hangul_string = hangul_string.replace("\\", "").replace("x","%").upper()[3:len(hangul_string)-1]
    hangul_string = "%EA%B9%80" + hangul_string #add the hangul string for Kim
    return hangul_string
```
